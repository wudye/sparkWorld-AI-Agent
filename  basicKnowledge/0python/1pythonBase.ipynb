```python
import time
import asyncio
```

```python
def make_coffee(customer:str) -> None:
    print(f"Making coffee for {customer}...")
    time.sleep(5)
    print("Coffee is ready!")
```

```python
def make_sync():
    start_time = time.time()
    make_coffee("Alice")
    make_coffee("Bob")
    make_coffee("Charlie")
    make_coffee("David")
    make_coffee("Eve")
    print(f"Total time taken: {time.time() - start_time:.2f} seconds")
```

```python
make_sync()
```

```python
async def make_coffee_async(customer:str) -> None:
    print(f"Making coffee for {customer}...")
    await asyncio.sleep(5)
    print("Coffee is ready!")

```

```python
async def make_async():
    start_time = time.time()
    tasks = [
        make_coffee_async("Alice"),
        make_coffee_async("Bob"),
        make_coffee_async("Charlie"),
        make_coffee_async("David"),
        make_coffee_async("Eve")
    ]
    await asyncio.gather(*tasks)
    print(f"Total time taken: {time.time() - start_time:.2f} seconds")
```

```python
# asyncio.run(make_async())
await make_async()
```
